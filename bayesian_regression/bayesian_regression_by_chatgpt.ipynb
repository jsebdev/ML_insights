```python
import numpy as np
import pymc3 as pm
import matplotlib.pyplot as plt

# Generate some random data
n = 100
x = np.linspace(0, 10, n)
y = 2 * x + np.random.normal(scale=1, size=n)

# Define the Bayesian regression model
with pm.Model() as model:
    # Define priors for the intercept and slope
    intercept = pm.Normal('intercept', mu=0, sd=10)
    slope = pm.Normal('slope', mu=0, sd=10)

    # Define the likelihood
    likelihood = pm.Normal('y', mu=intercept + slope*x, sd=1, observed=y)

    # Inference
    trace = pm.sample(1000, tune=1000)

# Plot the posterior distributions of the intercept and slope
pm.plot_posterior(trace, var_names=['intercept', 'slope'])
plt.show()
```
